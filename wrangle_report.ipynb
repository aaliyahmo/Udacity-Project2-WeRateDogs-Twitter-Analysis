### 3. Cleaning Data

> As the above picture states, data cleaning took the longest time. Everytime I thought I could move on and that its clean, I find something else. <br>
<br>
>I had to work on all the 9 issues and 4 tidy issues I found:<br>
<br>
> **Quality issues:**<br>
<br>
>1. Dataset contains tweets that are replies and not the rated tweets.<br>
    >This was resolved by dropping the rows that have values in the "in reply to" columns in twitter archive dataframe.<br>
    <br>
>2. The timestamp column is in string format and not in datetime format.<br>
    >Changing the column to datetime by using to_datetime().<br>
    <br>
>3. There are only 4 types of values in the source column, this can be simplified.<br>
    >Using value_counts() and then changing the 4 values into appropriate names to be easily understood. <br>
    <br>
>4. Dataset contains tweets that are retweets and not the rated tweets.<br>
    >Remove rows that are retweets by dropping rows that have values in the "retweeted_status_id" column.<br>
    <br>
>5. Some tweets have a denominator that isn't 10.<br>
    >The goal is to focus on ratings out of 10, so all tweets that were not 10 were dropped. However, the tweet were first checked for correct ratings before dropping, if it was incorrect and the correct rating was in the text, the rating was changed.<br>
    <br>
>6. Many dog names are "none", "a" or "an"<br>
    >Using a for loop to try to find the names after keywords such as "Here is".<br>
    <br>
>7. The tweet_ID column should be string and not int format.<br>
    >Changing the tweet_id column using astype() as this column is not a column calculations would be run on.<br>
    <br>
>8. Tweets with missing images.<br>
    >expanded_url is the column that gives a link to a image. No url means that there is no image and the row would be dropped.<br>
    <br>
>9. Keeping tweets thats a max 15 numerator.<br>
    >Removing outliers by keeping 15 the max number for the numerator.<br>
    <br>
<br>
>**Tidiness issues:** <br>
<br>
>1. Having one column for the 4 stages of dogs.<br>
    >The data started of with 4 columns, each column being a stage of dog (puppo, doggo, pupper and floofer), this is messy as usually theres only 1 type of dog stage per tweet. This would be simplified by creating 1 column called dog_stage.<br>
    <br>
>2. Remove purposeless (for our analysis) rows such as "in reply to" and "retweeted status".<br>
    > Dropping any unneccessary rows that will not be needed.<br>
    <br>
>3. Combine Json_tweets with df_archive. <br>
    >To compare the retweets and favourites with dog breeds, I had to join the dataframes as it would make it easier to analyze.<br>
><br>
>4. Adding image_predictions into df_archive.<br>
    >To compare the dog breeds that were recognized from the tweet images, I had to merge the 2 dataframes but first taking the dog breed with the highest confidence level.<br>
